```python
from glob import glob
import os

output_dir = "resources/data/healthy_preprocessed"
os.mkdir(output_dir)

imgs = glob("resources/data/healthy/*")
for i, path in enumerate(imgs):
    img = Image.open(path).rotate(90).resize((192,178))
    result = Image.new(img.mode, (256, 256), (0, 0, 0))
    result.paste(img, (32, 39))
    result.save(f"{output_dir}/{i}.png")
```
